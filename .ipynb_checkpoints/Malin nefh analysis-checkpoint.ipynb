```python
th= 2
nefh = adata.obs_vector("Nefh")>th
nav17 = adata.obs_vector("Scn9a")>th
trpv1 = adata.obs_vector("Trpv1")>th
trpm8 = adata.obs_vector("Trpm8")>th
piezo2 = adata.obs_vector("Piezo2")>th
```

```python
th= 0.5
flp = adata.obs_vector("Cntn1")>th
cre = adata.obs_vector("Lynx1")>th
```

```python
selection = np.logical_and(nefh, nav17)
selection = np.logical_and(np.logical_not(trpv1), selection)
selection = np.logical_and(np.logical_not(trpm8), selection)
selection = np.logical_and(np.logical_not(piezo2), selection)
```

```python
selection1 = np.logical_and(cre, flp)
```

```python
adata.obs["malinpop"] = ["Candidate UFN" if status == True else "rest" for status in selection]
adata.obs["creflp"] = ["Marked" if status == True else "untouched" for status in selection1]
```

```python
sc.pp.neighbors(adata, n_neighbors=15, n_pcs=9)
sc.tl.leiden(adata, resolution=1)
sc.tl.umap(adata)
sc.pl.umap(adata, color=["annotated", "malinpop","creflp", "Lynx1","Gap43", "leiden"], alpha = 1)
```
