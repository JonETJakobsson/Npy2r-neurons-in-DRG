# Analysis of Npy2r expression neurons in DRG scRNAseq experiments

This analysis is part of Ma and Gao *et al* 2019. Several datasets has been produced using scRNAseq on DRG tissue, here we will analyse cells from Li *et al* (GEO accession: **GSE63576**), Zeisel *et al* (**http://mousebrain.org/loomfiles_level_L6.html**) and Usoskin *et al* (**GSE59739**).

We want to investigate Npy2r expressing neurons, with relation to predefined functional cell types (Gatto *et al*) available to Usoskin and Zeisel datasets. To map these cell types onto the data, we will use Mutual nearest neighbour (MNN) algorithm (*Haghverdi et al. (2018), Batch effects in single-cell RNA-sequencing data are corrected by matching mutual nearest neighbors, Nature Biotechnology.*) to make the datasets comparable, then train a classifier (*Scikit-learn: Machine Learning in Python, Pedregosa et al., JMLR 12, pp. 2825-2830, 2011.*) to predict cell types on Li *et al* cells.

```python
import scanpy as sc
import pandas as pd
import numpy as np
import seaborn as sns

#Makes the text in figures editable in Illustrator
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['pdf.compression'] = 0
matplotlib.rcParams['ps.fonttype'] = 42
```

```python
#Load Usoskin DRG data
uso_data = pd.read_csv("Data/RPM_table.csv",
                       index_col=0, sep=";", decimal=",")
uso = sc.AnnData(uso_data.values.T)
uso.obs_names = uso_data.columns
uso.var_names = uso_data.index

# Load Meta data
meta = pd.read_csv("Data/metadata_cells.csv",
                   sep=";", index_col="Sample ID")
uso.obs = meta
good_cells = ["NP1", 'TH', 'NF5', 'NF2', 'NF3',
              'NP2', 'PEP2', 'NF4', 'PEP1', 'NF1', 'NP3']
keep = [True if cell_type in good_cells else False for cell_type in uso.obs["Level 3"]]
drg_u = uso[keep]

#add annotation from Gatto et al.
gatto = {"NF1":"Aδ-LTMR",
         "NF2":"Aβ-LTMR",
         "NF3":"Aβ-LTMR",
         "NF4":"Proprioceptor",
         "NF5":"Proprioceptor",
         "PEP1":"C-nociceptor",
         "PEP2":"Aδ-nociceptor",
         "NP1":"Mechanical MrgprD",
         "NP2":"Itch MrgprA3",
         "NP3":"Itch Nppb",
         "TH":"C-LTMR",
        }

drg_u.obs["annotated"] = [gatto[k] for k in drg_u.obs["Level 3"]]

qc = sc.pp.calculate_qc_metrics(drg_u)
sc.pp.normalize_per_cell(drg_u, counts_per_cell_after=np.median(np.apply_over_axes(np.sum, drg_u.X, axes=1)))
sc.pp.log1p(drg_u)
print(f"Average: {qc[0]['n_genes_by_counts'].mean()}\nSd: {qc[0]['n_genes_by_counts'].std()}")

drg_u_s = sc.pp.scale(drg_u, copy=True)
```

```python
# Load Zeisel DRG data
drg_z = sc.read_loom("Zeisel/l6_r3_peripheral_sensory_neurons.loom", sparse=False)
drg_z.var_names_make_unique()
drg_z.obs_names_make_unique()

#add annotation from Gatto et al.
gatto = {"PSNF1":"Aδ-LTMR",
         "PSNF2":"Aβ-LTMR",
         "PSNF3":"Proprioceptor",
         "PSPEP2":"C-nociceptor",
         "PSPEP3":"C-nociceptor",
         "PSPEP4":"C-nociceptor",
         "PSPEP5":"C-nociceptor",
         "PSPEP1":"Aδ-nociceptor",
         "PSNP2":"Mechanical MrgprD",
         "PSNP3":"Mechanical MrgprD",
         "PSNP4":"Itch MrgprA3",
         "PSNP5":"Itch MrgprA3",
         "PSNP6":"Itch Nppb",
         "PSNP1":"C-LTMR",
         "PSPEP6":"Cool",
         "PSPEP7":"Cool",
         "PSPEP8":"Cool"  
        }

drg_z.obs["annotated"] = [gatto[k] for k in drg_z.obs["ClusterName"]]


qc = sc.pp.calculate_qc_metrics(drg_z)
sc.pp.normalize_per_cell(drg_z, counts_per_cell_after=np.median(np.apply_over_axes(np.sum, drg_z.X, axes=1)))
sc.pp.log1p(drg_z)
print(f"Average: {qc[0]['n_genes_by_counts'].mean()}\nSd: {qc[0]['n_genes_by_counts'].std()}")

drg_z_s = sc.pp.scale(drg_z, copy=True)
```

```python
# Load Li et al data
li_data = pd.read_csv("Data/GSE63576_DRG_neurons_RNA_Seq.txt.gz", sep = "\\t", index_col = 0)
drg_l = sc.AnnData(li_data.values.T)
drg_l.obs_names = li_data.columns
drg_l.var_names = li_data.index


qc = sc.pp.calculate_qc_metrics(drg_l)
sc.pp.normalize_per_cell(drg_l, counts_per_cell_after=np.median(np.apply_over_axes(np.sum, drg_l.X, axes=1)))
sc.pp.log1p(drg_l)
print(f"Average: {qc[0]['n_genes_by_counts'].mean()}\nSd: {qc[0]['n_genes_by_counts'].std()}")

drg_l_s = sc.pp.scale(drg_l, copy=True)
```

# Merge datasets
Here we merge the three datasets using an outer merge. This will retain all genes detected in one of the datasets, and we set gene expression to zero for datasets that did not share this gene.

```python
merge_drg = drg_u.concatenate([drg_z, drg_l], batch_categories=["Usoskin", "Zeisel", "Li"], join="outer")
merge_drg.X = np.nan_to_num(merge_drg.X)
print((merge_drg.obs["batch"] == "Usoskin").sum(), "cells comes from Usoskin dataset")
print((merge_drg.obs["batch"] == "Zeisel").sum(), "cells comes from Zeisel dataset")
print((merge_drg.obs["batch"] == "Li").sum(), "cells comes from Li dataset")
```

Investigating the first 2 PC, we can se that differences between the datasets are captures in these PC.

```python
#Make color palette for all datasets
color3 = sns.palettes.color_palette("tab10", n_colors=3, desat=1)

sc.tl.pca(merge_drg)
sc.pl.pca_scatter(merge_drg, color="batch", palette=color3, alpha=0.5, save="_unmerged_and_unscaled.pdf")
```

# Data integration using MNN
Here we find the most variable genes in each dataset, and look for the inner join (intersect) of these genes. These genes are used in the MNN algorithm on scaled datasets.

```python
#filter HVG for datasets
sc.pp.filter_genes_dispersion(drg_u, n_top_genes=5000)
sc.pp.filter_genes_dispersion(drg_z, n_top_genes=5000)
sc.pp.filter_genes_dispersion(drg_l, n_top_genes=5000)
#drg_u = cn.genecall.filter_genes(drg_u)
#drg_z = cn.genecall.filter_genes(drg_z)
#Merging using inner join will keep only varables that are incommon
merge_inner = drg_u.concatenate([drg_z, drg_l], batch_categories=["Usoskin", "Zeisel", "Li"], join="inner")
#makes a list of shared HVG to use for MNN correction
join_genes = list(merge_inner.var_names)
print(len(join_genes))
```

```python
#Runs the MNN using only shared HVG in the scaled datasets
mnn_drg = sc.external.pp.mnn_correct(drg_u_s[:,join_genes], drg_z_s[:,join_genes], drg_l_s[:,join_genes], batch_categories = ["Usoskin", "Zeisel", "Li"], k = 10, save_raw = True)[0]
```

After the data integration step, the first two PCs are no longer describing variance between the datasets, but shared underlying variation.

```python
sc.tl.pca(mnn_drg)
sc.pl.pca_scatter(mnn_drg, color="batch", palette=color3, alpha=0.5, save="_merged_and_scaled.pdf")
```

Coloring the cells by dataset, we can see that all but the Cool cell cluster have representation of cells from each dataset.

# Cell type prediction
Li cells did not have any annotation metadata available, so here we train a linear SVC on cells with known annotations and gene profiles, and predict gene annotations for the Li cells.

```python
train_X = mnn_drg.X[mnn_drg.obs["batch"] != "Li", :]
train_y =mnn_drg.obs["annotated"][mnn_drg.obs["batch"] != "Li"].values
test_X =  mnn_drg.X[mnn_drg.obs["batch"] == "Li", :]
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_validate
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import ShuffleSplit

#evaluating linearSVC
clf_lsvm = LinearSVC(dual=True)
clf_lsvm.fit(train_X, train_y)

#using cross_val_score to estimate the overfitting and accuracy (Not working 20180917)
scores = cross_validate(estimator=clf_lsvm, X=train_X, y=list(train_y), cv=10)
print(f"Average score:\t {scores['test_score'].mean()} +- {scores['test_score'].std()} sd")

#evaluating Kneigbors
#clf_knc = KNeighborsClassifier(n_neighbors=5)
#clf_knc.fit(train_X, train_y)
#print(clf_knc.score(train_X, train_y))
```

```python
labels = clf_lsvm.predict(test_X)
#prob = clf_knc.predict_proba(mnn.X)
prob = clf_lsvm.decision_function(mnn_drg.X)
import numpy as np

prob = np.max(prob, axis = 1)
```

```python
mnn_drg.obs["annotated"][mnn_drg.obs["batch"] == "Li"] = labels
mnn_drg.obs["probability"] = prob
```

```python
sc.pl.umap(mnn_drg, color=["annotated"], palette=color, alpha = 0.5, save="_after_lable_prediction.pdf")
sc.pl.umap(mnn_drg, color=["probability"])
```
